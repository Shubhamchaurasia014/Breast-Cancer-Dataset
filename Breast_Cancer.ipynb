 <h1 font-size:40px;>Logistic Regression</h1>

```python
from sklearn.model_selection import train_test_split, cross_validate, StratifiedKFold
kfold= StratifiedKFold(n_splits=4)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, roc_auc_score
```

```python
x = df.drop(['target'], axis=1)
y = df[['target']]
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=32)
```

```python
lr = LogisticRegression(class_weight='balanced')
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```

<h1 font-size:40px;>Confusion Matrix</h1>

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(lr,x_train,y_train)
```

```python
cross_validate(lr, x, y, cv=4, scoring='recall', return_train_score=True, return_estimator=False)
```

```python
cross_validate(lr, x,y, cv=kfold, scoring='recall',return_train_score=True, return_estimator=False)
```

<h1 font-size:40px;>Decision Tree</h1>

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion='gini',
    splitter='best',
    max_depth=6,
    min_samples_split=3,
    min_samples_leaf=3,
    random_state=32,
    class_weight='balanced')
```

```python
cross_validate(dt, x, y, cv=kfold,   return_train_score=True,
    return_estimator=False)
```

<h1 font-size:40px;>Ensemble(Bagging) - Random Forest</h1>

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier( n_estimators=100,
    criterion='gini',
    max_depth=6,
    min_samples_split=3,
    min_samples_leaf=3,
    n_jobs= -1,
    random_state=32,
    class_weight='balanced')
```

```python
cross_validate(rfc, x, y, cv=kfold,  return_train_score=True,
    return_estimator=False )
```

<h1 font-size:40px;>Ensemble(Boosting) - Ada Boost</h1>

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
adb = AdaBoostClassifier(n_estimators=100,
    random_state=32)
```

```python
cross_validate(adb, x, y, cv=kfold,  return_train_score=True,
    return_estimator=True )
```

<h1 font-size:40px;>Hyperparametric Tuning on Random Forest by using GridSearchCV</h1>

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_rf = {'n_estimators':[50,100,150],
    'max_depth':[4,6,8],
    'min_samples_split':[3,4,6,8],
    'min_samples_leaf':[3,6,9],
    'class_weight':['balanced']   
}
```

```python
grid_search_rf = GridSearchCV(estimator=rfc,
    param_grid=param_grid_rf,
    scoring='recall',
    n_jobs=-1,
    cv=5)
```

```python
grid_search_rf.fit(x_train, y_train)
```

```python
best_model = grid_search_rf.best_estimator_
y_pred = best_model.predict(x_test)
```

```python
print(classification_report(y_pred,y_test))
```

<h1 font-size:40px;>Hyperparametric Tuning on Random Forest by using RandomizedSearchCV</h1>

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
param_rand_rf = {'n_estimators': randint(50,150),
    'max_depth': randint(4,8),
    'min_samples_split': randint(3,8),
    'min_samples_leaf': randint(3,9),
    'class_weight':['balanced']   
}
```

```python
random_search_rf = RandomizedSearchCV(estimator=rfc,
    param_distributions=param_rand_rf,
    n_iter=5,
    scoring='recall',
    n_jobs=-1,
    cv=kfold,
    random_state=36)
```

```python
random_search_rf.fit(x_train, y_train)
```

```python
best_model = random_search_rf.best_estimator_
y_pred = best_model.predict(x_test)
```

```python
print(classification_report(y_pred, y_test))
```
